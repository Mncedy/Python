```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Mnced\Documents\Python\Ice Cream Ratings.csv")
df
```

```python
df1 = df.set_index('Date')
```

```python
df.plot(kind = 'line')
```

```python
# Adding title and axis title
df.plot(kind = 'line', title = 'Sales', xlabel = 'Store', ylabel = 'Price' )
```

```python
# Separating line charts by each line
df1.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'bar')
```

```python
# Turning a bar graph into stacked
df.plot(kind = 'bar', stacked = True)
```

```python
# Looking at 1 column
df['Price'].plot(kind = 'bar') 
```

```python
# Horizontal bar chart (df1.plot.barh(stacked = true) :stacked horizontal)
df1.plot.barh()
```

```python
df1.plot.scatter(x = 'Price', y = 'Price')
```

```python
# Resizing plots
df1.plot.scatter(x = 'Price', y = 'Price', s = 200)
```

```python
# Changing colors
df1.plot.scatter(x = 'Flavor Rating', y = 'Flavor Rating', c = 'Red')
```

```python
# bins shows the distribution size of the bins
df1.plot.hist(bins= 4)
```

```python
df1.boxplot()
```

```python
df1.plot.area(figsize = (10,5))
```

```python
df1.plot.pie(y = 'Price',figsize = (10,7))
```
